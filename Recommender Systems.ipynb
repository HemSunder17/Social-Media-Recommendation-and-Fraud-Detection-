```python
# ================================
# 1. Import libraries
# ================================
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# ================================
# 2. Load dataset
# ================================
# replace with your file path
df = pd.read_csv('/Users/hemsunder/Desktop/GitHub Projects/RS_Project/Social_Media_Sentiment_Analysis_AI_Trends_2026.csv')

print("Dataset Loaded Successfully")
print(df.head())

# ================================
# 3. Create CATEGORY column
# ================================
def map_category(text):
    text = str(text).lower()
    
    if "car" in text:
        return "cars"
    elif "bike" in text or "motorcycle" in text:
        return "bikes"
    elif "motivation" in text or "fitness" in text:
        return "motivation"
    elif "dog" in text or "cat" in text:
        return "pets"
    else:
        return "other"

# change 'hashtags' to your actual column name if needed
df['category'] = df['hashtags'].apply(map_category)

# ================================
# 4. Create ACTUAL LABEL (liked)
# ================================
# modify column names if needed
df['liked'] = ((df['like_count'] + df['comment_count'] + df['share_count']) > 10).astype(int)

# ================================
# 5. Create PREDICTION (your system logic)
# ================================
# Example: user prefers cars and bikes
df['predicted'] = df['category'].apply(lambda x: 1 if x in ['cars', 'bikes'] else 0)

# ================================
# 6. Evaluation for SPECIFIC CATEGORY (cars)
# ================================
cars_df = df[df['category'] == 'cars']

print("\n===== Cars Category Evaluation =====")

if len(cars_df) > 0:
    y_true = cars_df['liked']
    y_pred = cars_df['predicted']
    
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred, zero_division=0))
    print("Recall:", recall_score(y_true, y_pred, zero_division=0))
    print("F1 Score:", f1_score(y_true, y_pred, zero_division=0))
else:
    print("No data available for cars category")

# ================================
# 7. Evaluation for ALL categories
# ================================
print("\n===== Category-wise Evaluation =====")

categories = df['category'].unique()

for cat in categories:
    temp = df[df['category'] == cat]
    
    if len(temp) == 0:
        continue
        
    print(f"\nCategory: {cat}")
    
    y_true = temp['liked']
    y_pred = temp['predicted']
    
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred, zero_division=0))
    print("Recall:", recall_score(y_true, y_pred, zero_division=0))
    print("F1 Score:", f1_score(y_true, y_pred, zero_division=0))

# ================================
# 8. Overall Evaluation (optional)
# ================================
print("\n===== Overall Model Performance =====")

print("Accuracy:", accuracy_score(df['liked'], df['predicted']))
print("Precision:", precision_score(df['liked'], df['predicted'], zero_division=0))
print("Recall:", recall_score(df['liked'], df['predicted'], zero_division=0))
print("F1 Score:", f1_score(df['liked'], df['predicted'], zero_division=0))
```
